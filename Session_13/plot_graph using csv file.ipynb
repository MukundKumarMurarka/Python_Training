```python
import pandas as pdf
import numpy as np
from matplotlib import pyplot as plt
```

```python

csv_file = pdf.read_csv("/home/mukund/Downloads/timedata.csv")
csv_data= pdf.DataFrame(csv_file)
csv_data.drop(["Unnamed: 7","Unnamed: 8","Unnamed: 9"], inplace =True, axis =1)
csv_data.head(5)
```

```python
#1. plot a graph 
from matplotlib import pyplot as plt
x= list(csv_data.iloc[:,2])
y = list(csv_data.iloc[:,5])
plt.xlabel("open")
plt.ylabel("colse")
plt.plot(x,y ,"g", )
plt.show()
```

```python
# bar graph :- 
x1= list(csv_data.iloc[:,0])
y1=list(csv_data.iloc[:,1])
plt.bar(x1, y1, color='g')
plt.title("samsung prices on the basis of month")
plt.xlabel("Date")
plt.ylabel("open prices")
plt.show()
```

```python
# Scatter plot :- 
x= list(csv_data.iloc[:,1])
y = list(csv_data.iloc[:,4])

plt.scatter(x,y, label ="skitscat", color ="k")
plt.title("pie plot")
plt.show()
```
